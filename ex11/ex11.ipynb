```python
img = cv2.imread('apple2.jpeg')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(img_grey, 230, 255, cv2.THRESH_BINARY)
thresh = cv2.bitwise_not(thresh)
contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[-2]

img = cv2.drawContours(img_rgb, contours, -1, (0,255,0), 2)

plt.imshow(img)
#plt.imshow(thresh, cmap='gray')


```

```python
c = max(contours, key=cv2.contourArea)
((x, y), radius) = cv2.minEnclosingCircle(c)

img_circle = img_rgb.copy()

cv2.circle(img_circle, center=(int(x), int(y)), radius=int(radius), color=(255, 255, 0), thickness=10)

plt.imshow(img_circle)
    
```
