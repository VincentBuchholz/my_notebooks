````python
import numpy as np
filename = 'befkbhalderstatkode.csv'

bef_stats = np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)
````

````python
neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave', 
          5: 'Valby', 6: 'Vanløse', 7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst', 
          10: 'Amager Vest', 99: 'Udenfor'}
````

Find out how many people lived in each of the 11 areas in 2015

````python
year = bef_stats[:,0] == 2015

for code in neighb:
    neighborhood = bef_stats[:,1] == code
    amount = np.sum(bef_stats[year & neighborhood][:,4])
    
    print(f'{amount} lived in {neighb[code]} in 2015')
    
    
````

Make a bar plot to show the size of each city area from the smallest to the largest in 2015

````python
year = bef_stats[:,0] == 2015

city_area_size={}

for code in neighb:
    neighborhood = bef_stats[:,1] == code
    amount = np.sum(bef_stats[year & neighborhood][:,4])
    city_area_size[neighb[code]]=amount
    

city_area_size_sorted = dict(sorted(city_area_size.items(), key=lambda item: item[1]))

import matplotlib.pyplot as plt
plt.bar(list(city_area_size_sorted.keys()),list(city_area_size_sorted.values()),width=0.5, align='center')
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
````

Create a boolean mask to find out how many people above 65 years lived in Copenhagen in 2015

````python
year = bef_stats[:,0] == 2015
age = bef_stats[:,2] <= 65
mask = (year & age)
np.sum(bef_stats[mask][:,4])
````

How many of those were from the other nordic countries (not dk). Hint: see notebook: "04 Numpy"

````python
year = bef_stats[:,0] == 2015
age = bef_stats[:,2] <= 65
statecode = bef_stats[:,3] != 5100
mask = (year & age & statecode)
np.sum(bef_stats[mask][:,4])
````

 Make a line plot showing the changes of number of people in vesterbro and østerbro from 1992 to 2015

````python
import matplotlib.pyplot as plt
city_part = bef_stats[:,1]
years = np.unique(bef_stats[:,0])

pop1 = np.array([np.sum(bef_stats[(bef_stats[:,1] == 2) & (bef_stats[:,0] == year)][:,4]) for year in years])
pop2 = np.array([np.sum(bef_stats[(bef_stats[:,1] == 4) & (bef_stats[:,0] == year)][:,4]) for year in years])
    

plt.plot(years,pop1,years,pop2)
````

````python
import pandas as pd 
url = 'https://api.statbank.dk/v1/data/FOLK1A/CSV?valuePresentation=Code&delimiter=Semicolon&OMR%C3%85DE=000%2C084%2C147%2C400%2C085%2C083%2C082%2C081%2C851%2C461%2C561%2C751&K%C3%98N=1%2C2&ALDER=0%2C1%2C2%2C3%2C4%2C5%2C6%2C7%2C8%2C9%2C10%2C11%2C12%2C13%2C14%2C15%2C16%2C17%2C18%2C19%2C20%2C21%2C22%2C23%2C24%2C25%2C26%2C27%2C28%2C29%2C30%2C31%2C32%2C33%2C34%2C35%2C36%2C37%2C3%2C39%2C40%2C41%2C42%2C43%2C44%2C45%2C46%2C47%2C48%2C49%2C50%2C51%2C52%2C53%2C54%2C55%2C56%2C57%2C58%2C59%2C60%2C61%2C62%2C63%2C64%2C65%2C66%2C67%2C68%2C69%2C70%2C71%2C72%2C73%2C74%2C75%2C76%2C77%2C78%2C79%2C80%2C81%2C82%2C83%2C84%2C85%2C86%2C87%2C88%2C89%2C90%2C91%2C92%2C93%2C94%2C95%2C96%2C97%2C98%2C99%2C100&Tid=2008K1%2C2009K1%2C2010K1%2C2011K1%2C2012K1%2C2013K1%2C2014K1%2C2015K1%2C2016K1%2C2017K1%2C2018K1%2C2019K1%2C2020K1'
df = pd.read_csv(url,sep=';')
df['TID'] = df['TID'].map(lambda x:x[:-2]) #cut the last 2 characters
df.to_csv('demografic_cleaned.csv',header=False, index=False)
````

````python
dd = np.loadtxt('demografic_cleaned.csv', delimiter=',', dtype=np.uint)
````

````python
import csv

with open('DKstat_bykoder.csv','r') as file:
    reader = csv.reader(file)
    cities = {int(rows[0].split(';')[0]):rows[0].split(';')[1] for rows in reader}
````

Now find out which city in 2020 has the highest percentage of men (gender code 1)

````python
def summed_by_city(city_code):
    area = dd[:,0] == city_code
    male = dd[:,1] == 1
    year = dd[:,3] == 2020
    
    male_mask = (area & male & year)
    
    males = np.sum(dd[male_mask][:,4])
    amount = np.sum(dd[area & year][:,4])
    
    if(males == 0 or amount == 0):
        return 0
    
    return (males/amount)*100
    
    
    
````

````python
summed_by_city(561)
````

````python
highest_male_pct = {'city': '','pct_male':0}
for city_code in cities:
    pct = summed_by_city(city_code)
    if pct > highest_male_pct['pct_male']:
        highest_male_pct['city'] = cities[city_code]
        highest_male_pct['pct_male'] = pct
highest_male_pct
    
````

Create a line chart with 5 lines showing population changes in the 5 largest cities from 2008 to 2020

````python
#københavn 84
#Aarhus 751
#Odense 461
#Aalborg 851
#Esbjerg 561

import matplotlib.pyplot as plt
years = np.unique(dd[:,3])

cph = np.array([np.sum(dd[(dd[:,3] == year) & (dd[:,0] == 84)][:,4]) for year in years])
aar = np.array([np.sum(dd[(dd[:,3] == year) & (dd[:,0] == 751)][:,4]) for year in years])
ode = np.array([np.sum(dd[(dd[:,3] == year) & (dd[:,0] == 461)][:,4]) for year in years])
aal = np.array([np.sum(dd[(dd[:,3] == year) & (dd[:,0] == 851)][:,4]) for year in years])
esb = np.array([np.sum(dd[(dd[:,3] == year) & (dd[:,0] == 561)][:,4]) for year in years])

plt.plot(years, cph, label = "København")
plt.plot(years, aar, label = "Aarhus")
plt.plot(years,ode, label ="Odense")
plt.plot(years,aal, label ="Aalborg")
plt.plot(years,esb, label ="Esbjerg")
plt.legend()
plt.show()
````
